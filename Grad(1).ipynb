```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from rouge import Rouge

tokenizer = AutoTokenizer.from_pretrained("nsi319/legal-led-base-16384")
model = AutoModelForSeq2SeqLM.from_pretrained("nsi319/legal-led-base-16384")

padding = "max_length"

rouge = Rouge()
all_scores = []
for i in range(min(1, len(train_data))):
    example = train_data[i]


    text = f"""{example['sources']}"""
    input_tokenized = tokenizer.encode(text, return_tensors='pt', padding=padding, pad_to_max_length=True, max_length=6144, truncation=True)


    summary_ids = model.generate(
        input_tokenized,
        num_beams=8,
        no_repeat_ngram_size=3,
        length_penalty=2,
        min_length=200,
        max_length=450
    )


    generated_summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True, clean_up_tokenization_spaces=False)

    summaryLong=multi_lexsum1['validation'][i]['summary/long']


    reference_summary = summaryLong


    scores = rouge.get_scores(generated_summary, reference_summary)
    all_scores.append(scores)

    print(generated_summary)

    ##print(f"Document {i + 1} - Rouge Scores: {scores}")

##output_file_path = '/content/drive/MyDrive/Rouge-n Scores/rouge_scores.txt'
##with open(output_file_path, 'w') as txt_file:
    ##for i, scores in enumerate(all_scores, start=1):
        ##txt_file.write(f"Document {i} - Rouge-N Scores: {scores}\n")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from rouge import Rouge

tokenizer = AutoTokenizer.from_pretrained("nsi319/legal-pegasus")
model = AutoModelForSeq2SeqLM.from_pretrained("nsi319/legal-pegasus")

padding = "max_length"

rouge = Rouge()
all_scores = []
for i in range(min(50, len(train_data))):  # Iterate up to the first 50 examples
    example = train_data[i]


    text = f"""{example['sources']}"""
    input_tokenized = tokenizer.encode(text, return_tensors='pt',max_length=1024,truncation=True)
    summary_ids = model.generate(input_tokenized,
                                  num_beams=9,
                                  no_repeat_ngram_size=3,
                                  length_penalty=2.0,
                                  min_length=150,
                                  max_length=250,
                                  early_stopping=True)
    summary = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=False) for g in summary_ids][0]


    summaryLong=multi_lexsum1['validation'][i]['summary/long']


    reference_summary = summaryLong


    scores = rouge.get_scores(summary, reference_summary)
    all_scores.append(scores)


    print(f"Document {i + 1} - Rouge Scores: {scores}")

output_file_path = '/content/drive/MyDrive/Rouge-n Scores Model2/rouge_scores.txt'
with open(output_file_path, 'w') as txt_file:
    for i, scores in enumerate(all_scores, start=1):
        txt_file.write(f"Document {i} - Rouge-N Scores: {scores}\n")
```
